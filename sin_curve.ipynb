```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def sin_curve(x, A, f, p):
    return A * np.sin(f * x + p)
def cos_curve(x, A, f, p):
    return A * np.cos(f * x + p)
def tan_curve(x, A, f, p):
    return A * np.tan(f * x + p)

```

```python
x = np.linspace(0, 2 * np.pi, 100)
noise = np.random.normal(0, 0.5, x.shape)
y = sin_curve(x, 1, 3, 0)+noise #Setting Y values with noise 
plt.figure(figsize=(10, 10))
plt.plot(x, y, label='sin(x)', color='blue')

opt, cov = curve_fit(sin_curve, x, y, [1, 3, 0])#predict initial values

plt.plot(x, sin_curve(x, *opt), label='fit', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Sine Curve Fit')
```

```python
x = np.linspace(0, 2 * np.pi, 100)
noise = np.random.normal(0, 0.3, x.shape)
y = cos_curve(x, 1, 3, 0)+noise #Setting Y values with noise
plt.figure(figsize=(10, 10))
plt.plot(x, y, label='cos(x)', color='blue')
opt, cov = curve_fit(cos_curve, x, y, [1, 3, 0])#predict initial values
plt.plot(x, cos_curve(x, *opt), label='fit', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Cosine Curve Fit')
```

```python
x = np.linspace(0, 2 * np.pi, 100)
noise = np.random.normal(0, 1, x.shape)
y = tan_curve(x, 1, 3, 0)+noise #Setting Y values with noise
plt.figure(figsize=(10, 10))
plt.plot(x, y, label='tan(x)', color='blue')
opt, cov = curve_fit(tan_curve, x, y, [1, 3, 0])#predict initial values
plt.plot(x, tan_curve(x, *opt), label='fit', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Tangent Curve Fit')
```

```python
x = np.linspace(0, 2 * np.pi, 200)
noise = np.random.normal(0, 10, x.shape)
y = (9*x+7)+noise #Setting Y values with noise 
plt.figure(figsize=(10, 10))
plt.scatter(x, y, label='sin(x)', color='blue')

slope, b = np.polyfit(x, y, 1)#predict initial values
y2 = slope * x + b
plt.plot(x, y2, label='fit', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Sine Curve Fit')

x_means = np.mean(x)
y_means = np.mean(y)
numerator = np.sum((x - x_means) * (y - y_means))
denominator = np.sqrt(np.sum((x - x_means)**2)) * np.sqrt(np.sum((y - y_means)**2))
r = numerator/denominator
plt.text(max(x)-0.6, min(y), f'R: {r:.5f}', fontsize=12, ha='left', va='center')
```
